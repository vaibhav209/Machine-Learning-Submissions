```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv(r'Bodyfat.csv')
df
```

```python
df.isnull().sum()
```

```python
#replace the missing values
mean = df['bodyfat'].mean()
df['bodyfat'].fillna(mean, inplace=True)

median1 = df['Age'].median()
df['Age'].fillna(median1, inplace=True)

mean2 = df['Weight'].mean()
df['Weight'].fillna(mean2, inplace=True)

mean3 = df['Height'].mean()
df['Height'].fillna(mean3, inplace=True)

mean4 = df['Neck'].mean()
df['Neck'].fillna(mean4, inplace=True)

mean5 = df['Chest'].mean()
df['Chest'].fillna(mean5, inplace=True)

mean6 = df['Abdomen'].mean()
df['Abdomen'].fillna(mean6, inplace=True)

mean7 = df['Hip'].mean()
df['Hip'].fillna(mean7, inplace=True)
```
